```python
import numpy as np
import matplotlib.pyplot as plt
def frac(T):
    E1 = 13.6/1**2;
    E2 = 13.6/2**2;
    k = 1.38 * 10 ** -23/1.602*10**19;
    f = np.exp(-E1/k/T)/np.exp(-E2/k/T);
    return f
print(frac(5000))    
```

```python
def ionfrac(T):
    Eion = 13.6/1**2*1.602*10**-19;
    k=1.38 * 10 ** -23;
    h=6.626*10**-34;
    f = 2*(2*np.pi*9.11*10**-31*k*T/h**2)**(3/2)*np.exp(-Eion/k/T)/10**20;
    return f
print(ionfrac(5000))
```

```python
def neutral1frac(T):
    ne=10**20;
    neutral = frac(T);
    ion=ionfrac(T);
    neutral1=neutral/(1+ion);
    return neutral1
print(neutral1frac(5000))
```

```python
T=np.linspace(100,50000,10)
firsttoground = frac(T)
ion=ionfrac(T)
firsttoall=neutral1frac(T)
plt.plot(T,firsttoground)
plt.legend(['fraction of 1st excited to ground', 'ion fraction', 'fraction of 1st to all'])
plt.xlabel('Temperature')
plt.ylabel('fraction')
```

```python

plt.plot(T,ion)
plt.legend(['fraction of 1st excited to ground', 'ion fraction', 'fraction of 1st to all'])
plt.xlabel('Temperature')
plt.ylabel('fraction')
```

```python
plt.plot(T,firsttoall)
plt.legend(['fraction of 1st to all'])
plt.xlabel('Temperature')
plt.ylabel('fraction')
```
