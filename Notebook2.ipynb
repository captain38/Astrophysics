```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math as math
```

```python
x = np.zeros(20)
x[0:5] = 10
x[5:15] = np.arange(12,31,2)
x[15:]= 30
plt.plot(x)
plt.plot([4, 4], [8, 32],'k--',color = 'red')
plt.plot([14, 14], [8, 32],'k--',color = 'red')
plt.ylim(8,32)
```

```python
x = np.array([[4, 2, 3, 2], [2, 4, 3, 1], [2, 4, 1, 3], [4, 1, 2, 3]])
print(x)
print('The first row of x is')
print(x[0][0:4])
```

```python
print('The first column of x is ')
print(x[:,0])
print('The 3rd row of x is ')
print(x[2])
print('The last 2 columns of x is ')
print(x[:,-2:])
print('The top right corner of x is ')
print(x[:2,2:])
print('The middle 4 values of x are ')
print(x[1:3,1:3])
```

```python
x=np.linspace(0,20,101)
y=np.sin(x)
plt.plot(x[y > 0],y[y > 0], 'bo')
plt.plot(x[y<=0],y[y<=0],'ro')
```

```python
arr = np.loadtxt('xypoints.dat')
x=arr[0,:]
y=arr[1,:]
plt.plot(x,y,'ro')
plt.plot(x[(x < -2) & (-5 <= y) & (y < 0)],y[(x < -2) & (-5 <= y) & (y < 0)],'bo')
plt.plot(x[(((x-5)**2+y**2) <25)],y[(((x-5)**2+y**2) <25)],'go')
plt.axis('equal')
```
