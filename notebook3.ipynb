```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
months = ['January','Febuary','March','April','May','June','July','August','September','October','November','December']
daysInMonth = [31,28,31,30,31,30,31,31,30,31,30,31]
for i in range(12):
    print('There are ', daysInMonth[i], ' days in ', months[i])
print('End of year')
```

```python
x=np.linspace(-2*np.pi,2*np.pi,100)
y=np.zeros(100)
for i in range(100):
    if(x[i] < 0):
        y[i]=np.cos(x[i])
    else:
        y[i]=np.exp(-x[i])
plt.plot(x,y)
plt.xlim(-2*np.pi,2*np.pi);
```

```python
holland = np.loadtxt('holland_temperature.dat')
for i in range(12):
    if (holland[i] >= 10):
        print('In ', months[i], ' the average temperature was above 10 degrees')
    else:
        print('In ', months[i], ' the average temperature was not above 10 degrees')
```

```python
data = np.array([1, 3, 2, 5, 7, 3, 4, 2])
datasum = np.zeros(len(data))
datasum[0] = data[0]
for i in range(1,len(data)):
    datasum[i]= datasum[i-1] + data[i]
print('data:', data,'\npartialsum:',datasum)
print('Sum with numpy:', np.cumsum(data))
```

```python
difFrom15 = abs(holland - 15)
minimum = difFrom15[0]
minIndex = 0
for i in range(len(difFrom15)):
    if (difFrom15[i] <= minimum):
        minimum = difFrom15[i]
        minIndex = i
print('The closest temperature is ', holland[minIndex], ' degrees\nin month ', months[minIndex])
altmin = np.argmin(abs(holland - 15))
print('Alternative method yields ', holland[altmin], 'degrees\nin month',months[minIndex])
```

```python
oil = np.loadtxt('oil_price.dat',delimiter=',')
plt.plot(oil[:,2])
nrow,ncol=oil.shape
flag_40 = False
flag_60 = False
index40=0
index60=0
index80=0
i=0
while (i < len(oil[:,2])):
    
    if(oil[i,2]>40):
        if(flag_40==False):
            flag_40=True
            index40=i
        if(oil[i,2]>60):
            if(flag_60==False):
                flag_60=True;
                index60=i
            if(oil[i,2]>80):
                index80=i
                
                break
    i+=1
print('The price exceeds 40 euros in ',months[int(oil[index40,1])], ' ', oil[index40,0], ' at ', oil[index40,2])
print('The price exceeds 60 euros in ',months[int(oil[index60,1])], ' ', oil[index60,0], ' at ', oil[index60,2])
print('The price exceeds 80 euros in ',months[int(oil[index80,1])], ' ', oil[index80,0], ' at ', oil[index80,2])
```
