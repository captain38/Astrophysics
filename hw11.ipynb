```python
import numpy as np
import math
import matplotlib.pyplot as plt
def distanceFinder(e,theta):
    r = np.zeros(len(theta))
    for t in np.arange(len(theta)):
        if(1-e*math.cos(theta[t]!=0)):
            r[t]=1/(1-e*math.cos(theta[t]))
        else:
            r[t]=1e20#Eliminates div by 0 and maintains size of array r.
    return r
```

```python
theta = np.arange(0,2*math.pi,.02*math.pi)
```

```python
ep = [0,.5,.9,1.0,2.0,10.0]
```

```python

r=[[0] * len(theta) for i in np.arange(len(ep))]
fig, ax = plt.subplots(figsize=(20, 10))
ax = plt.subplot(1,len(ep),1,projection='polar')
for i in np.arange(len(ep)):
    ax = plt.subplot(1,len(ep),i+1,projection='polar')
    ax.set_rlim(0,10)
    r[i]=distanceFinder(ep[i],theta)
    ax.plot(theta,r[i])
```

```python
def orbVel(e,theta):
    v=np.zeros(len(theta))
    for i in np.arange(len(theta)):
        v[i]=-e*math.sin(theta[i])/(1-e*math.cos(theta[i]))**2*distanceFinder(e,[theta[i]])
    return v
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
axV=plt.subplot(1,len(ep),1,projection='polar')
for i in np.arange(len(ep)-3):
    axV=plt.subplot(1,len(ep),i+1,projection='polar')
    axV.set_rlim(0,10)
    axV.plot(theta,orbVel(ep[i],theta))
```
